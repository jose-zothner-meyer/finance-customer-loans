```python
import pandas as pd

# Load the dataset from the CSV file
df = pd.read_csv('loan_payments.csv')

# Display the first few rows to inspect the data
df.head()
```

```python
df['term']
```

```python
df['term'].str.replace('months', '', regex=False).astype(float)
```

```python
import numpy as np

df['employment_length'] = df['employment_length'].replace({
            '10+ years': '10',
            '< 1 year': '0.5',
            'n/a': np.nan
        })
```

```python
# Remove ' years' or ' year' from the strings
df['employment_length'] = df['employment_length'].str.replace('years', '', regex=False)
df['employment_length'] = df['employment_length'].str.replace('year', '', regex=False)
```

```python
# Convert the column to float
df['employment_length'] = df['employment_length'].astype(float)
```
